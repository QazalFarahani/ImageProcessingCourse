# Enhancement

we want to increase the brightness of dark pixels in the image. the method we will be using is log function. log function due to its shape map the darker pixels to a larger range. so that the difference of dark pixels increase and the image looks brighter.

$$\frac{255 \log (1 + \alpha x)}{\log (1 + 255 \alpha)}$$
